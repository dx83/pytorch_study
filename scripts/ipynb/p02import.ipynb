```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, TensorDataset
from torch.optim.lr_scheduler import ReduceLROnPlateau
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import seaborn as sns

print(f"PyTorch version: {torch.__version__}")
print(f"XPU available: {torch.xpu.is_available()}")

torch.manual_seed(42)
np.random.seed(42)
```

```python
def perceptron_and(x1, x2):
    w = torch.tensor([0.5, 0.5])    # 가중치
    b = torch.tensor(-0.7)          # 편향
    x = torch.tensor([x1, x2], dtype=torch.float32)
    ouput = torch.dot(w, x) + b     # 가중치 합 연산
    return 1 if ouput >= 0 else 0

print("AND 게이트 테스트:")
print(f"perceptron_and(0, 0) = {perceptron_and(0, 0)}")
print(f"perceptron_and(0, 1) = {perceptron_and(0, 1)}")
print(f"perceptron_and(1, 0) = {perceptron_and(1, 0)}")
print(f"perceptron_and(1, 1) = {perceptron_and(1, 1)}")
```
