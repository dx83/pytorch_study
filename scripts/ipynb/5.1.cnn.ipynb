```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import torchvision.models as models
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image
import os   # 소문자
import time
import seaborn as sns
from sklearn.metrics import confusion_matrix
from torchsummary import summary
import torch.nn.utils.prune as prune

device = torch.device("xpu" if torch.xpu.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# 기존 심층 신경망의 이미지 처리
class SimpleDNN(nn.Module):
    def __init__(self):
        super(SimpleDNN, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(28 * 28, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 10)
    
    def forward(self, x):
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# 모델 생성 및 테스트
dnn_model = SimpleDNN()
print("SimpleDNN 모델 구조:")
print(dnn_model)

# 테스트 입력
test_input = torch.randn(1, 1, 28, 28)
dnn_output = dnn_model(test_input)
print(f"DNN 출력 크기: {dnn_output.shape}")
```

```python
# 파라미터 공유
sobel_filter = torch.tensor([[[
    [1.0, 0.0, -1.0],
    [2.0, 0.0, -2.0],
    [1.0, 0.0, -1.0]
]]])

sample_28x28 = torch.randn(1, 1, 28, 28)
sobel_output = F.conv2d(sample_28x28, sobel_filter)

print(f"Sobel 필터 출력 크기: {sobel_output.shape}")
```
