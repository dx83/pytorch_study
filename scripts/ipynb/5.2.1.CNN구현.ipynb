```python
import torch
import torch.nn as nn

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # 첫 번째 합성곱
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, stride=1, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        # 두 번째 합성곱
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        # 완전 연결 레이어
        self.fc = nn.Linear(2048, 10)

    def forward(self, x):
        # 첫 번째 블록             입력 3 x 32 x 32
        x = self.conv1(x)          # 16 x 32 x 32
        x = self.relu1(x)
        x = self.pool1(x)          # 16 x 16 x 16
        # 두 전째 블록
        x = self.conv2(x)          # 32 x 16 x 16
        x = self.relu2(x)
        x = self.pool2(x)          # 32 x  8 x  8
        # 평탄화
        x = x.view(x.size(0), -1)  # 32 x  8 x  8 = 2048
        # 완전 연결 레이어
        x = self.fc(x)             # 출력 10
        return x

# 모델 인스턴스 생성
model = SimpleCNN()
print(model)
# 임의의 입력으로 모델 테스트
sample_input = torch.randn(1, 3, 32, 32)
output = model(sample_input)
print(f"출력 크기: {output.shape}")  # [1, 10]
```

```python
# 배치 정규화 적용 합성곱 신경망
class CNNWithBatchNorm(nn.Module):
    def __init__(self):
        super(CNNWithBatchNorm, self).__init__()
        # 첫 번째 합성곱
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(16)   # 배치 정규화 추가
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        # 두 번째 합성곱
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(32)   # 배치 정규화 추가
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        # 완전 연결 레이어
        self.fc = nn.Linear(2048, 10)

    def forward(self, x):
        # 첫 번째 블록
        x = self.conv1(x)
        x = self.bn1(x)     # 배치 정규화 적용
        x = self.relu1(x)
        x = self.pool1(x)
        # 두 번째 블록
        x = self.conv2(x)
        x = self.bn2(x)     # 배치 정규화 적용
        x = self.relu2(x)
        x = self.pool2(x)
        # 평탄화
        x = x.view(x.size(0), -1)
        # 완전 연결 레이어
        x = self.fc(x)
        return x

# 모델 생성 및 테스트
model = CNNWithBatchNorm()
print(model)
# 더미 입력으로 테스트
dummy_input = torch.randn(1, 3, 32, 32)
output = model(dummy_input)
print(f"출력 크기: {output.shape}")
```

```python
import torch
import torch.nn as nn
# 드롭아웃 적용 합성곱 신경망
class CNNWithDropout(nn.Module):
    def __init__(self):
        super(CNNWithDropout, self).__init__()
        # 첫 번째 합성곱
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(16)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout1 = nn.Dropout2d(0.25)  # 25% 드롭아웃 적용
        # 두 번째 합성곱
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(32)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout2 = nn.Dropout2d(0.25)  # 25% 드롭아웃 적용
        # 완전 연결 레이어
        self.fc1 = nn.Linear(32 * 8 * 8, 128)
        self.dropout3 = nn.Dropout2d(0.5)   # 50% 드롭아웃 적용
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        # 첫 번째 블록
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu1(x)
        x = self.pool1(x)
        x = self.dropout1(x)    # 드롭아웃 적용
        # 두 번째 블록
        x = self.conv2(x)
        x = self.bn2(x)
        x = self.relu2(x)
        x = self.pool2(x)
        x = self.dropout2(x)    # 드롭아웃 적용
        # 평탄화
        x = x.view(x.size(0), -1)
        # 완전 연결 레이어
        x = self.fc1(x)
        x = self.dropout3(x)    # 드롭아웃 적용
        x = self.fc2(x)
        return x

# 모델 생성 및 출력
model = CNNWithDropout()
print(model)
```
