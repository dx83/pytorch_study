```python
def pizza_price_model(size):
    weight = 2
    bias = 3
    price = weight * size + bias
    return price
```

```python
size = 5
predicted_price = pizza_price_model(size)
print(f"{size}인치 피자 예상 가격: {predicted_price}천원")
```

```python
import torch
def forward_pass(x, W, b):  # 입력 텐서, 가중치 행렬, 바이어스
    z = torch.mm(x, W) + b  # 선형 변환: 가중치 곱셈과 바이어스 추가
    output = torch.relu(z)  # 활성화 함수 적용: ReLU를 사용하여 비선형성 추가
    return output

x = torch.tensor([[1.0, 2.0],
                  [3.0, 4.0]])

W = torch.tensor([[0.5, -0.5],
                  [1.0, 1.0]])

b = torch.tensor([[1.0, 1.0]])

output = forward_pass(x, W, b)
output
```

```python
# 간단한 성형 회귀 모델
import torch
import torch.nn as nn

x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y_true = 2 * x + 1

# 임의의 모델 예측값
model_output = torch.tensor([[2.5], [3.5], [4.5], [5.5]])
# 손실 함수
criterion = nn.MSELoss()
# 손실 계산
loss = criterion(model_output, y_true)
print(f"손실 값 (MSE): {loss.item()}")
```

```python
size = 5

actual_price = 11
predicted_price = 13
error = predicted_price - actual_price
print(f"오차: {error}처원 (너무 비싸게 예측)")
print("가중치를 조금 줄여야겠다...")

weight = 2
bias = 3

learning_rate = 0.01
weight = weight - learning_rate * error * size
bias = bias - learning_rate * error
print(f"새로운 가중치: {weight:.2f}, 새로운 편향: {bias:.2f}")
```
